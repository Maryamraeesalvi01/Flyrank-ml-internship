## 3. Quick look at the data (2–3 real numbers)

The starter dataset contains **30,000 pages**, making manual review impractical. Among these pages, **16,262 are trending down**, suggesting that a large number of pages may require attention but cannot all be reviewed individually. Additionally, average engagement rates differ across trend directions, with up-trending pages showing higher average engagement **(2.99)** than flat pages **(1.37)**. These observations support the need for a data-driven approach to prioritize content optimization opportunities.

```python
import pandas as pd
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")
print(df.shape)
df.head()


print(f"Total pages: {len(df):,}")
print(df["trend_direction"].value_counts())
print(df.groupby("trend_direction")["engagement_rate"].mean().round(2))


import matplotlib.pyplot as plt
df["trend_direction"].value_counts().plot(kind="bar")
plt.title("Distribution of Trend Direction")
plt.xlabel("Trend Direction")
plt.ylabel("Number of Pages")

plt.show()
```
